```
import torch
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

if torch.cuda.is_available():
  device=torch.device("cuda")
else:
  device=torch.device("cpu")

n_epochs=20
batch_size=25
learning_rate=0.1

transform=transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))])
train_dataset=torchvision.datasets.CIFAR10(root="./data",train=True,transform=transform,download=True)
test_dataset=torchvision.datasets.CIFAR10(root="./data",train=False,transform=transform,download=True)
train_loader=torch.utils.data.DataLoader(dataset=train_dataset,batch_size=batch_size,shuffle=True)
test_loader=torch.utils.data.DataLoader(dataset=test_dataset,batch_size=batch_size,shuffle=False)

map_labels={0:"plane",1:"car",2:"bird",3:"cat",4:"deer",5:"dog",6:"frog",7:"horse",8:"ship",9:"truck"}
examples=iter(train_loader)
samples,labels=examples.next()
print(samples.shape,labels.shape)
images=samples.numpy()

for i in range(25):
  plt.subplot(5,5,i+1)
  img_view=np.moveaxis(images[i],0,2)
  plt.imshow(img_view)

for i in range(25):
  print(map_labels[labels[i].item()])

print(torch.cuda.device_count())
print(torch.cuda.get_device_name(0))
```

```
import torch.nn.functional as F

class ResNet50(nn.Module):
  def __init__(self):
    super(ResNet50,self).__init__()
    self.conv1a=nn.Conv2d(3,96,3)
    self.bn1a=nn.BatchNorm2d(96)
    self.conv1b=nn.Conv2d(96,96,3,padding=1)
    self.bn1b=nn.BatchNorm2d(96)
    self.conv1c=nn.Conv2d(96,96,3,padding=1)
    self.bn1c=nn.BatchNorm2d(96)
    self.conv1d=nn.Conv2d(96,96,3,padding=1)
    self.bn1d=nn.BatchNorm2d(96)
    self.conv1e=nn.Conv2d(96,96,3,padding=1)
    self.bn1e=nn.BatchNorm2d(96)
    self.conv1f=nn.Conv2d(96,96,3,padding=1)
    self.bn1f=nn.BatchNorm2d(96)
    self.conv1g=nn.Conv2d(96,96,3,padding=1)
    self.bn1g=nn.BatchNorm2d(96)
    self.mp1=nn.MaxPool2d(2,2)
    self.conv2a=nn.Conv2d(96,192,3,padding=1)
    self.bn2a=nn.BatchNorm2d(192)
    self.conv2b=nn.Conv2d(192,192,3,padding=1)
    self.bn2b=nn.BatchNorm2d(192)
    self.conv2c=nn.Conv2d(192,192,3,padding=1)
    self.bn2c=nn.BatchNorm2d(192)
    self.conv2d=nn.Conv2d(192,192,3,padding=1)
    self.bn2d=nn.BatchNorm2d(192)
    self.conv2e=nn.Conv2d(192,192,3,padding=1)
    self.bn2e=nn.BatchNorm2d(192)
    self.conv2f=nn.Conv2d(192,192,3,padding=1)
    self.bn2f=nn.BatchNorm2d(192)
    self.conv2g=nn.Conv2d(192,192,3,padding=1)
    self.bn2g=nn.BatchNorm2d(192)
    self.conv2h=nn.Conv2d(192,192,3,padding=1)
    self.bn2h=nn.BatchNorm2d(192)
    self.conv2i=nn.Conv2d(192,192,3,padding=1)
    self.bn2i=nn.BatchNorm2d(192)
    self.conv2j=nn.Conv2d(192,192,3,padding=1)
    self.bn2j=nn.BatchNorm2d(192)
    self.conv2k=nn.Conv2d(192,192,3,padding=1)
    self.bn2k=nn.BatchNorm2d(192)
    self.conv2l=nn.Conv2d(192,192,3,padding=1)
    self.bn2l=nn.BatchNorm2d(192)
    self.conv2m=nn.Conv2d(192,192,3,padding=1)
    self.bn2m=nn.BatchNorm2d(192)
    self.mp2=nn.MaxPool2d(2,2)
    self.conv3a=nn.Conv2d(192,256,3,padding=1)
    self.bn3a=nn.BatchNorm2d(256)
    self.conv3b=nn.Conv2d(256,256,3,padding=1)
    self.bn3b=nn.BatchNorm2d(256)
    self.conv3c=nn.Conv2d(256,256,3,padding=1)
    self.bn3c=nn.BatchNorm2d(256)
    self.conv3d=nn.Conv2d(256,256,3,padding=1)
    self.bn3d=nn.BatchNorm2d(256)
    self.conv3e=nn.Conv2d(256,256,3,padding=1)
    self.bn3e=nn.BatchNorm2d(256)
    self.conv3f=nn.Conv2d(256,256,3,padding=1)
    self.bn3f=nn.BatchNorm2d(256)
    self.conv3g=nn.Conv2d(256,256,3,padding=1)
    self.bn3g=nn.BatchNorm2d(256)
    self.conv3h=nn.Conv2d(256,256,3,padding=1)
    self.bn3h=nn.BatchNorm2d(256)
    self.conv3i=nn.Conv2d(256,256,3,padding=1)
    self.bn3i=nn.BatchNorm2d(256)
    self.conv3j=nn.Conv2d(256,256,3,padding=1)
    self.bn3j=nn.BatchNorm2d(256)
    self.conv3k=nn.Conv2d(256,256,3,padding=1)
    self.bn3k=nn.BatchNorm2d(256)
    self.conv3l=nn.Conv2d(256,256,3,padding=1)
    self.bn3l=nn.BatchNorm2d(256)
    self.conv3m=nn.Conv2d(256,256,3,padding=1)
    self.bn3m=nn.BatchNorm2d(256)
    self.mp3=nn.MaxPool2d(2,2)
    self.conv4a=nn.Conv2d(256,256,3)
    self.conv4b=nn.Conv2d(256,192,1)
    self.conv4c=nn.Conv2d(192,10,1)
    
  def forward(self,x):
    conv1a=self.bn1a(F.relu(self.conv1a(x)))
    conv1b=self.bn1b(F.relu(self.conv1b(conv1a)))
    conv1c=self.bn1c(F.relu(self.conv1c(conv1b)))
    conv1d=self.bn1d(F.relu(self.conv1d(conv1c)))+conv1a
    conv1e=self.bn1e(F.relu(self.conv1e(conv1d)))
    conv1f=self.bn1f(F.relu(self.conv1f(conv1e)))+conv1c
    conv1g=self.bn1g(F.relu(self.conv1g(conv1f)))
    mp1=self.mp1(conv1g+conv1e)
    conv2a=self.bn2a(F.relu(self.conv2a(mp1)))
    conv2b=self.bn2b(F.relu(self.conv2b(conv2a)))
    conv2c=self.bn2c(F.relu(self.conv2c(conv2b)))
    conv2d=self.bn2d(F.relu(self.conv2d(conv2c)))+conv2a
    conv2e=self.bn2e(F.relu(self.conv2e(conv2d)))
    conv2f=self.bn2f(F.relu(self.conv2f(conv2e)))+conv2c
    conv2g=self.bn2g(F.relu(self.conv2g(conv2f)))
    conv2h=self.bn2h(F.relu(self.conv2h(conv2g)))+conv2e
    conv2i=self.bn2i(F.relu(self.conv2i(conv2h)))
    conv2j=self.bn2j(F.relu(self.conv2j(conv2i)))+conv2g
    conv2k=self.bn2k(F.relu(self.conv2k(conv2j)))
    conv2l=self.bn2l(F.relu(self.conv2l(conv2k)))+conv2i
    conv2m=self.bn2m(F.relu(self.conv2m(conv2l)))
    mp2=self.mp2(conv2m+conv2k)
    conv3a=self.bn3a(F.relu(self.conv3a(mp2)))
    conv3b=self.bn3b(F.relu(self.conv3b(conv3a)))
    conv3c=self.bn3c(F.relu(self.conv3c(conv3b)))
    conv3d=self.bn3d(F.relu(self.conv3d(conv3c)))+conv3a
    conv3e=self.bn3e(F.relu(self.conv3e(conv3d)))
    conv3f=self.bn3f(F.relu(self.conv3f(conv3e)))+conv3c
    conv3g=self.bn3g(F.relu(self.conv3g(conv3f)))
    conv3h=self.bn3h(F.relu(self.conv3h(conv3g)))+conv3e
    conv3i=self.bn3i(F.relu(self.conv3i(conv3h)))
    conv3j=self.bn3j(F.relu(self.conv3j(conv3i)))+conv3g
    conv3k=self.bn3k(F.relu(self.conv3k(conv3j)))
    conv3l=self.bn3l(F.relu(self.conv3l(conv3k)))+conv3i
    conv3m=self.bn3m(F.relu(self.conv3m(conv3l)))
    mp3=self.mp3(conv3m+conv3k)
    conv4a=F.relu(self.conv4a(mp3))
    conv4b=F.relu(self.conv4b(conv4a))
    conv4c=F.relu(self.conv4c(conv4b))
    out=torch.mean(conv4c.view(conv4c.size(0),conv4c.size(1),-1),dim=2)
    return out

model=ResNet50()

loss=nn.CrossEntropyLoss()
optimizer=torch.optim.SGD(model.parameters(),lr=learning_rate)


for epoch in range(n_epochs):
  for i,(features,labels) in enumerate(train_loader):
    labels_pred=model(features)
    l=loss(labels_pred,labels)
    
    l.backward()
    optimizer.step()
    optimizer.zero_grad()
    
    if(i%10==9):
      print(epoch,i,l)
  
  Filename="model_"+str(epoch)+".pth"
  torch.save(model.state_dict(),Filename)

  with torch.no_grad():
    n_samples=0
    n_correct=0
    for features,labels in test_loader:
      pred=model(features)
      _,predictions=torch.max(pred,1)
      n_correct+=(predictions==labels).sum().item()
      n_samples+=labels.shape[0]
    acc=100.0*n_correct/n_samples
    print(acc)
```
